```python
import numpy as np
import pandas as pd
import warnings 
%matplotlib inline
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
from sklearn import preprocessing
warnings.filterwarnings('ignore')
```

```python
#导入数据
#读取去零数据
#Caco-2	CYP3A4	hERG	HOB	MN
data_train = pd.read_excel('Molecular_Descriptor_delzero.xlsx',engine='openpyxl',index_col=0)
data_test = pd.read_excel('Molecular_Descriptor.xlsx',engine='openpyxl',sheet_name='test',index=True)
data_label= pd.read_excel('ADMET.xlsx',index_col=0)
```

```python
data_tests = pd.DataFrame()
for index,col in data_train.items():
    data_tests[index] = data_test[index]
```

```python
data_train.drop(['SMILES'],axis=1,inplace=True)
data_tests.drop(['SMILES'],axis=1,inplace=True)
```

```python
#test去0
# data_train = data_train.loc[:,(data_train!=0).any(axis=0)]
# data_test = data_test.loc[:,(data_test!=0).any(axis=0)]
```

```python
#归一化
m_scale = preprocessing.MinMaxScaler()
data_train = pd.DataFrame(m_scale.fit_transform(data_train))
data_tests = pd.DataFrame(m_scale.fit_transform(data_tests))
```

```python
#网格搜索f1_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,KFold,cross_val_score
from sklearn.svm import SVC
from xgboost import XGBClassifier
import joblib
def rmsle_cv(model,X,y):
    acc = cross_val_score(model,X,y,scoring='accuracy',cv=5)
    return acc
class grid():
    def __init__(self,model):
        self.model = model
    def grid_get(self,X,y,parm_grid):
        grid_search = GridSearchCV(self.model,param_grid,cv=5,scoring='accuracy')
        grid_search.fit(X,y)
        print(grid_search.best_params_,grid_search.best_score_)
#         grid_search.cv_results_['mean_test_score'] = grid_search.cv_results_['mean_test_score']
#         print(pd.DataFrame(grid_search.cv_results_)[['params','mean_test_score','std_test_score']])
print("Starting===========================================")
datasets = ['Caco-2','CYP3A4','hERG','HOB','MN']
# param_grid = {
#     'penalty':['l1','l2','elasticnet','none'],'max_iter':[200,500,1000]}
# LR = LogisticRegression()
# param_grid = {
#     'C':[2,3,5],'kernel':['poly'],
#     'degree':[4,5,6],'gamma':['scale'],'max_iter':[100,300,500],
#     'random_state':[10]
# }
# param_set = {'C': 3, 'degree': 4, 'gamma': 'scale', 'kernel': 'poly', 'max_iter': 300, 'random_state': 10}
LR = XGBClassifier()
param_grid = {
    'gamma':[0.1],
    'min_child_weight':[1,3],
    'max_depth':[4,6,8],
    'n_estimators':[100,300]
}
for i in datasets:
    opt_LR = GridSearchCV(estimator=LR,param_grid=param_grid,scoring='f1',cv=5,verbose=1,n_jobs = 64)
    X_train, X_test, Y_train, Y_test = train_test_split(data_train, data_label[i], test_size=0.3)
    opt_LR.fit(X_train,Y_train)
#     joblib.dump(opt_LR, 'xgboost_{}_.pkl'.format(i))
    print('{}类别\n每轮输入日志：{}\n参数的最佳取值:{}\n最佳模型得分:{}'.format(i,opt_LR.cv_results_,opt_LR.best_params_,opt_LR.best_score_))
```

```python
#PRE
Caco = joblib.load('xgboost_Caco-2_.pkl')
res_1 = pd.DataFrame(Caco.predict(data_tests))
res_1.columns = ['Caco-2']
CYP = joblib.load('xgboost_CYP3A4_.pkl')
res_2 = pd.DataFrame(CYP.predict(data_tests))
res_2.columns = ['CYP3A4']
hE = joblib.load('xgboost_hERG_.pkl')
res_3 = pd.DataFrame(hE.predict(data_tests))
res_3.columns = ['hERG']
HOB = joblib.load('xgboost_HOB_.pkl')
res_4 = pd.DataFrame(HOB.predict(data_tests))
res_4.columns = ['HOB']
MN = joblib.load('xgboost_MN_.pkl')
res_5 = pd.DataFrame(MN.predict(data_tests))
res_5.columns = ['MN']
results = pd.concat([res_1,res_2,res_3,res_4,res_5],axis=1)
```
